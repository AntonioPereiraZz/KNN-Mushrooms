### Convertendo  Object  Data em Numeric Data

```python
mushrooms = pd.get_dummies(df,columns=['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat'])
```

```python
mushrooms.head(n=7)
```

```python
mushrooms.dtypes
```

### Dividir os Dados em Treino e Teste

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(mushrooms.drop(columns=['class']), mushrooms['class'], test_size=0.3)
```

### Instnciando o algoritimo KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

### Treinando Algoritimo 

```python
knn.fit(X_train,y_train)
```

### Executando o KNN

```python
resultado = knn.predict(X_test)
```

```python
resultado
```

# Validao

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test,resultado))
```

```python
# Podemos identificar que mesmo utilizando o parmetro K com o valor igual a 5, este Script de treinamento KNN obteve uma preciso de 100% quando comparamos o resultado da predio com a poro do y_test
```

# Otimizao do parmetro K

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_list = list(range(1,31))
```

```python
k_values = dict(n_neighbors=k_list)
```

### Instanciar objeto GridSearch

```python
grid = GridSearchCV(knn, k_values, cv=5, scoring='accuracy')
```

### Treinando Objeto

```python
grid.fit(mushrooms.drop(columns=['class']),mushrooms['class'])
```

### Visualizando os Valores

```python
grid.cv_results_['mean_test_score']
```

```python
grid.best_params_
```

```python
scores = []
for mean in grid.cv_results_['mean_test_score']:
    scores.append(mean)
```

```python
plt.figure(figsize=(8,4))
plt.plot(k_list,scores,color='blue',linestyle='-',marker='o')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.show
```
